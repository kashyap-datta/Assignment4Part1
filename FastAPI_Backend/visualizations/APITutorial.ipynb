```python
import zipfile
from PIL import Image
import matplotlib.pyplot as plt

def image_gen():
    imgzip = zipfile.ZipFile("archive.zip")
    inflist = imgzip.infolist()

    for f in inflist:
        ifile = imgzip.open(f)
        img = Image.open(ifile)
        #img.show()
        #print(img)
        plt.figure(figsize= (12,12))
        plt.imshow(img)

image_gen()
```

```python
import requests
data ={
  "location": "STONEHAM",
  "begintime": "01-AUG-19 14:46:00",
  "endtime": "01-AUG-19 14:46:00"
}

res = requests.post('http://127.0.0.1:8000/predict',json=data)
image_gen()
```
